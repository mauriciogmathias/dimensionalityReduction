```python
import requests
from PIL import Image
import matplotlib.pyplot as plt
```

```python
url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png';
```

```python
#getting the image
data = requests.get(url).content;

#creating file img with extension .png
f = open('colored_img.png','wb');

#storing the image data inside the data variable to the file
f.write(data);
f.close();

#opening the saved image and displaying it
img = Image.open('colored_img.png');
plt.imshow(img);
```

```python
#extracting pixel map
pixel_map = img.load();

#extracting the width and height
width, height = img.size;

#Doing the grayscale first

#looping through all image pixels
for i in range(width):
    for j in range(height):

        #getting the RGB pixel value
        r, g, b = img.getpixel((i, j));

        #apply formula of grayscale
        grayscale = (0.299*r + 0.587*g + 0.114*b);

        #setting the pixel value
        pixel_map[i, j] = (int(grayscale), int(grayscale), int(grayscale));

#saving final grayscale output
img.save("grayscale_img", format="png");
plt.imshow(img);
```

```python
#looping through all image pixels
for i in range(width):
    for j in range(height):

        #getting the RGB pixel value
        pixel = img.getpixel((i, j));

        #verifying pixel intensity
        if pixel[0] > 130:
          pixel_map[i, j] = (255,255,255);
        else:
          pixel_map[i, j] = (0,0,0);

#saving final grayscale output
img.save("binary_img", format="png");
plt.imshow(img);
```
